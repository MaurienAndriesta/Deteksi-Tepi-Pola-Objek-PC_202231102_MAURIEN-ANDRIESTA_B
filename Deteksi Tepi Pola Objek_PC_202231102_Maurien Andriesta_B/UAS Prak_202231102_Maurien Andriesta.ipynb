```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Membaca gambar dari file
img = cv2.imread('maurinn.jpg')

# Mengubah ukuran gambar menjadi tinggi 1400 piksel
height = 1400
aspect_ratio = height / img.shape[0]
new_width = int(img.shape[1] * aspect_ratio)
img_resized = cv2.resize(img, (new_width, height))

# Konversi gambar ke warna RGB
img_rgb = cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB)

# Membuat variabel edges untuk menampung hasil canny (deteksi tepi) dari variabel img
edges = cv2.Canny(img_resized, 100, 150)

# Menggunakan tampilan matplotlib
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Sebagai fungsi numpy untuk konversi array menjadi flat array
ax = axs.ravel()

# Tampilan asli
ax[0].imshow(img_rgb)
ax[0].set_title("Original Image")

# Tampilan edges
ax[1].imshow(edges, cmap='gray')
ax[1].set_title("Canny Edge Detection")

# Menampilkan hasil
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Membaca gambar dari file
img = cv2.imread('maurinn.jpg')

# Mengubah ukuran gambar menjadi tinggi 1400 piksel
height = 1400
aspect_ratio = height / img.shape[0]
new_width = int(img.shape[1] * aspect_ratio)
img_resized = cv2.resize(img, (new_width, height))

# Konversi gambar ke warna RGB
img_rgb = cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB)

# Membuat variabel edges untuk menampung hasil canny (deteksi tepi) dari variabel img
edges = cv2.Canny(img_resized, 100, 150)

# Menemukan kontur dalam peta tepi
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Membuat salinan dari gambar asli untuk menggambar kontur
contour_image = img_resized.copy()

# Menggambar kontur pada salinan gambar
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)

# Menggunakan tampilan matplotlib
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Sebagai fungsi numpy untuk konversi array menjadi flat array
ax = axs.ravel()

# Tampilan asli
ax[0].imshow(img_rgb)
ax[0].set_title("Original Image")
ax[0].axis('off')

# Tampilan edges
ax[1].imshow(edges, cmap='gray')
ax[1].set_title("Canny Edge Detection")
ax[1].axis('off')

# Tampilan kontur
ax[2].imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
ax[2].set_title("Contours Detection")
ax[2].axis('off')

# Menampilkan hasil
plt.show()
```
